```python
import pandas as pd

data = pd.read_csv("results/results0_statistical_analysis.csv", skipinitialspace=True, header=0)
```

```python
import warnings
warnings.filterwarnings(
    action='ignore', category=UserWarning, message=r"Boolean Series.*"
)

queries = []
for key in data.QUERY.unique():
    bm25 = data[:][data.QUERY == key][data.NAME == "BM25"]["ALPHA-NDCG"].values[0]
    fr = data[:][data.QUERY == key][data.NAME == "FR"]["ALPHA-NDCG"].values[0]
    mmr = data[:][data.QUERY == key][data.NAME == "MMR"]["ALPHA-NDCG"].values[0]
    cl = data[:][data.QUERY == key][data.NAME == "CL"]["ALPHA-NDCG"].values[0]
    queries.append(
        (key,
        {
            "bm25":bm25,
            "fr":fr,
            "mmr":mmr,
            "cl":cl
        }
        )
    )

# for key in data.QUERY.unique():
#     bm25 = data[:][data.QUERY == key][data.NAME == "BM25"]["TREC"].values[0]
#     fr = data[:][data.QUERY == key][data.NAME == "FR"]["TREC"].values[0]
#     mmr = data[:][data.QUERY == key][data.NAME == "MMR"]["TREC"].values[0]
#     cl = data[:][data.QUERY == key][data.NAME == "CL"]["TREC"].values[0]
#     queries.append(
#         (key,
#         {
#             "bm25":bm25,
#             "fr":fr,
#             "mmr":mmr,
#             "cl":cl
#         }
#         )
#     )
    
print(queries)
```

```python

# Importing library 
import scipy.stats as stats 


# sample data
pre = [30, 31, 34, 40, 36, 35, 
       34, 30, 28, 29] 
  
post = [30, 31, 32, 38, 32, 31, 
        32, 29, 28, 30] 
  
# Performing the paired sample t-test 
print(stats.ttest_rel(pre, post)) 
```

```python
import numpy as np
```

```python
bm25 = [x["bm25"] for q,x in queries]
fr = [x["fr"] for q,x in queries]
mmr = [x["mmr"] for q,x in queries]
cl = [x["cl"] if not np.isnan(x["cl"]) else 0 for q,x in queries]

print(stats.ttest_rel(bm25, mmr))  
print(stats.ttest_rel(fr, mmr)) 
print(stats.ttest_rel(cl, mmr)) 
```
